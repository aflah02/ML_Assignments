### 1) Making The Dataset

## 3) PTA Training

#### Training on Noise Free Data

```python
pta = Perceptron(epochs=1000, min_error_threshold=0.01)
```

```python
pta.train(data, verbose=1)
```

#### Decision Boundary for Noise Free Data

```python
pta.plot_decision_boundary(data, 'Perceptron Decision Boundary')
```

#### Training for Noisy Data

```python
pta2 = Perceptron(epochs=1000, min_error_threshold=0.01)
```

```python
pta2.train(noisy_data, verbose=1)
```

#### Decision Boundary for Noisy Data

```python
pta2.plot_decision_boundary(noisy_data, 'Perceptron Decision Boundary')
```
